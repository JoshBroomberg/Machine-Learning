```python
from PIL import Image
import PIL.ImageOps    
from glob import glob

from PIL import ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True

import numpy as np
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')

SEED = 3
np.random.seed(seed=SEED)
```

### Process Images

Resize the images, flatten each into a pixel vector, and globally-center the pixel vector by removing the mean.

```python
NUM_IMAGES_PER_GENDER = 500
IMAGE_SIZE = (120, 150)

male_images = np.random.choice(glob('original_images/man_images/*'),
                        NUM_IMAGES_PER_GENDER, replace=False)
female_images = np.random.choice(glob('original_images/woman_images/*'),
                     NUM_IMAGES_PER_GENDER, replace=False)

images = []

def read_image(image_file):
    with open(image_file, 'r+b') as f:
        with Image.open(f) as image:
            resized = image.resize(IMAGE_SIZE)
            return np.array(resized).flatten()

# Process images, creating labelled pixel vectors
# which will be useful in a classifier.
images = [(read_image(path), 1, path) for path in male_images] +\
    [(read_image(path), 0, path) for path in female_images]

# Prepare for PCA by separating and centering the pixel data.
raw_image_data = np.array([image[0] for image in images])
class_data = np.array([image[1] for image in images]) # useful for LDA later.

# Centering
global_pixel_means = np.mean(raw_image_data, axis=0) 
image_data = raw_image_data - global_pixel_means

image_data.shape
```

## PCA

### Run PCA without Component Constraint

First, run the PCA without a specified number of components. This is useful for exploring how many components we really need.

From the plot, it looks like the first 8 components do most of the explaining. After this, we get a very small (and roughly constant) from each component. However, these 8 don't explain much in absolute terms (~0.65). This indicates our data is not easily compressible, which is somewhat natural given how rudemntry the cleaning was.

```python
pca = PCA()
pca.fit(image_data, whiten=True, copy=True, random_state=SEED)
differences = np.ediff1d(pca.explained_variance_ratio_)
summed = np.cumsum(pca.explained_variance_ratio_)
```

```python
plt.figure(figsize=(10, 10))
plot_num = 50

plt.scatter(
    list(range(plot_num)), pca.explained_variance_ratio_[:plot_num], 
    color='green', label='Explained variance proportion')

plt.scatter(list(range(plot_num)), differences[:plot_num], 
    color='blue', label='Delta in explanatory power')

plt.scatter(list(range(plot_num)), summed[:plot_num], 
    color='red', label='Total explanatory power')

plt.legend(loc=1, fontsize=16)
plt.xlabel(r'Component number', size=16)
plt.ylabel(r'% variance explained', size=16)
plt.show()
```

### Repeat PCA with N=8

Run the PCA with out desired number of components.

```python
pca = PCA(n_components=50, whiten=True, copy=True, random_state=SEED)
pca.fit(image_data)
```

Additionally, get the transformed (compressed) data set.

```python
compressed_image_data = pca.transform(image_data)
uncompressed_image_data = pca.inverse_transform(compressed_image_data)
```

### Analyze the Results

We know explore what our components seem to be showing.

```python
# Normalize a pixel vector and return
# an image. This code is from:
# https://github.com/joelgrus/shirts/blob/master/visuals.py#L55
def image_from_component_values(component):
    """takes one of the principal components and turns it into an image"""
    hi = max(component)
    lo = min(component)
    n = int(len(component) / 3)
    divisor = hi - lo
    if divisor == 0:
        divisor = 1
    
    def rescale(x):
        return int(255 * (x - lo) / divisor)
    
    d = [(rescale(component[3 * i]),
          rescale(component[3 * i + 1]),
          rescale(component[3 * i + 2])) for i in range(n)]
    im = Image.new('RGB',IMAGE_SIZE)
    im.putdata(d)
    return im
```

```python
fig=plt.figure(figsize=(12, 20))
columns = 4
rows = 8
for i in range(1, 33, 4):
    component_id = int(i/4)
    eigen_image = image_from_component_values(pca.components_[component_id])
    inverted_egein_image = PIL.ImageOps.invert(eigen_image)
    
    most_similar_id = (np.argmax(compressed_image_data[:, component_id]))
    most_dissimilar_id = (np.argmin(compressed_image_data[:, component_id]))

    
    fig.add_subplot(rows, columns, i)
    plt.imshow(eigen_image)
    
    fig.add_subplot(rows, columns, i+1)
    similar_image = plt.imshow(
        raw_image_data[most_similar_id, :].reshape(IMAGE_SIZE[1], IMAGE_SIZE[0], 3))
    
    fig.add_subplot(rows, columns, i+2)
    similar_image = plt.imshow(
        raw_image_data[most_dissimilar_id, :].reshape(IMAGE_SIZE[1], IMAGE_SIZE[0], 3))

    
    fig.add_subplot(rows, columns, i+3)
    plt.imshow(inverted_egein_image)
    
plt.show()

# Light to dark center
# Black vs white closes
# Light to dark top to bottom
# Light dark left to right
# light sleeves vs dark sleeves
# red vs blue
# dark-light rings
# dark vs light pocket area
```

## LDA

```python
# Do a test/train split on the raw data.
X_train, X_test, y_train, y_test = train_test_split(image_data, class_data, test_size=0.2, 
                                                    stratify=class_data, random_state=SEED)
```

### Classification without Dimension Reduction

```python
parameters = {'C': [10 ** x for x in range(5)],
             'gamma': [10 ** (-1 * x) for x in range(5)]}

svm = SVC(random_state=SEED)
svm_cv = GridSearchCV(svm, parameters, n_jobs=-1)
svm_cv.fit(X_train, y_train)

print('Train accuracy:', sum(svm_cv.predict(X_train) == y_train) / X_train.shape[0])
print('Test accuracy:', sum(svm_cv.predict(X_test) == y_test) / X_test.shape[0])
print(classification_report(svm_cv.predict(X_test), y_test))
```

### Use LDA to reduce dimenions

```python
lda = LinearDiscriminantAnalysis(n_components=1, solver="svd")
lda.fit(X_train, y_train)

print("Variance explained: {}".format(np.sum(lda.explained_variance_ratio_)))

print('Train accuracy:', sum(lda.predict(X_train) == y_train) / X_train.shape[0])

print("Test accuracy")
print(classification_report(lda.predict(X_test), y_test))
```

```python
lda_x_train = lda.transform(X_train)
lda_x_test = lda.transform(X_test)
```

```python
parameters = {'C': [10 ** x for x in range(5)],
             'gamma': [10 ** (-1 * x) for x in range(5)]}

lda_svm = SVC(random_state=SEED)
lda_svm_cv = GridSearchCV(lda_svm, parameters, n_jobs=-1)
lda_svm_cv.fit(lda_x_train, y_train)

print('Train accuracy:', sum(lda_svm_cv.predict(lda_x_train) == y_train) / lda_x_train.shape[0])
print('Test accuracy:')
print(classification_report(lda_svm_cv.predict(lda_x_test), y_test))
```

### Use PCA to reduce dimensions

```python
clf_pca = PCA(n_components=50, whiten=True)
clf_pca.fit(X_train)
clf_pca.components_.shape
```

```python
pca_x_train = clf_pca.transform(X_train)
pca_x_test = clf_pca.transform(X_test)
```

```python
parameters = {'C': [10 ** x for x in range(5)],
             'gamma': [10 ** (-1 * x) for x in range(5)]}

pca_svm = SVC(random_state=SEED)
plc_svm_cv = GridSearchCV(pca_svm, parameters, n_jobs=-1)
plc_svm_cv.fit(pca_x_train, y_train)

print('Train accuracy:', sum(plc_svm_cv.predict(pca_x_train) == y_train) / pca_x_train.shape[0])
print('Test accuracy:')
print(classification_report(plc_svm_cv.predict(pca_x_test), y_test))
```
