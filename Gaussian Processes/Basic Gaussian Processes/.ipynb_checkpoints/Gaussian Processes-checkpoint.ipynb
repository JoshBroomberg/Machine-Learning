```python
import GPy
import numpy as np
import pylab as pb
import matplotlib.pyplot as plt
from time import sleep
```

```python
X = np.random.uniform(-3.,3.,(1,1))
Y = np.sin(X) + np.random.randn(20,1)*0.05
```

```python
kernel = GPy.kern.RBF(input_dim=1, variance=1., lengthscale=1.)
m = GPy.models.GPRegression(X,Y,kernel)
```

```python
m
```

```python
fig = m.plot()
GPy.plotting.show(fig)
```

```python
m.optimize(messages=True)
```

```python



# X = X[:,None] # reshape X to make it n*D
mu = np.zeros((20)) # vector of the means
C = kernel.K(X,X) # covariance matrix

# Generate 20 sample path with mean mu and covariance C
Z = np.random.multivariate_normal(mu,C,2)

pb.figure() # open new plotting window
for i in range(2):
    print(Z[i,:])
    pb.scatter(X.flatten(), Y.flatten())
    pb.scatter(sorted(X.flatten()),Z[i,:])
    
    


```

```python
x_space = np.linspace(-5, 5, 100).reshape(-1, 1)

rand_x = lambda: np.random.uniform(-3.,3.,(1,1))
rand_y = lambda x: np.sin(x) + np.random.randn(1,1)*0.05

X = rand_x()
y = rand_y(X)

kernel = GPy.kern.RBF(input_dim=1, variance=1., lengthscale=1.)
m = GPy.models.GPRegression(X,y,kernel)
    
for i in range(1,10):
    m.optimize(messages=False)
    
    posteriorTestY = m.posterior_samples_f(x_space, full_cov=True, size=10)
    
    fig = m.plot()
    GPy.plotting.show(fig)
    
#     plt.scatter(X.flatten(), y.flatten())
    plt.plot(x_space, posteriorTestY)
    plt.show()
    
    new_x = rand_x()
    new_y = rand_y(new_x)
    
    X = np.vstack([X, new_x])
    y = np.vstack([y, new_y])
    
    m.set_XY(X=X, Y=y)
    
    sleep(1)
    
```
